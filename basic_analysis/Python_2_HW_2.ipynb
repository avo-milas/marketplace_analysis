## Задание 1 (10 баллов)

Определите:

* Число товаров

* Среднюю стоимость товара (стоимость товара = среднее от цен в датасете items)

в разрезе категорий (все категории должны быть на английском языке)

Табличка, которая у вас должна получиться:

```
category | products | price

value    | value      | value
```

**Обратите внимание:**

Для категории portateis_cozinha_e_preparadores_de_alimentos перевод portable kitchen and food preparers

Для категории pc_gamer перевод PC Gamer

Для них нужно отдельно добавить перевод

````python
%%write_and_run task_1.py

# важно! все зависимости, которые используете (если добавляее новые) в этом классе надо явно продублировать в эту ячейку
import numpy as np
import pandas as pd

def task_1(translation, items, products):
    to_add = pd.DataFrame({"product_category_name": ["portateis_cozinha_e_preparadores_de_alimentos", "pc_gamer"],
                        "product_category_name_english" : ["portable kitchen and food preparers", "PC Gamer"]})
    translation = pd.concat([translation, to_add], ignore_index = True)
    res = products.merge(translation, on='product_category_name', how='left')
    res.drop('product_category_name', axis=1, inplace=True)

    items1 = items.groupby(['product_id']).agg({'price': 'mean'})
    items1.reset_index(inplace=True)
    res = res.merge(items1, on='product_id', how='left')

    agg_funcs = {
            'price' : 'mean', 'product_id' : 'count'
    }
    res = res.groupby(['product_category_name_english']).agg(agg_funcs)
    res.reset_index(inplace=True)

    res.rename(columns={'product_category_name_english': 'category', 'product_id': 'products'}, inplace=True)

    return res
````

````python
#Проверки

res = task_1(translation, items, products)

assert res[res.category == 'portable kitchen and food preparers'].price.values[0] == 186.996
assert len(res) == 73
assert len(res.drop_duplicates()) == 73
assert res[res.category == 'drinks'].products.values[0] == 81
assert res.products.sum() == 32341
assert res.price.sum() == 12459.751444351941
assert res[res.category == 'home_confort'].price.values[0] == 185.56926417326417
````

## Задание 3 (10 баллов)

Выведите долю потраченных денег по каждому штату (потраченные деньги - сумма денег по доставленным заказам, сумма по price и freight_value)

*Примечание:* разбивка по штатам - по штату покупателя, процент - число от 0 до 1


Табличка, которая у вас должна получиться:

```
state | perc

value | value
```

````python
%%write_and_run task_3.py

# важно! все зависимости, которые используете (если добавляее новые) в этом классе надо явно продублировать в эту ячейку
import numpy as np
import pandas as pd

def task_3(orders, customers, items):
    grouped_ord = items.groupby(['order_id'], as_index=False).agg({'price' : 'sum', 'freight_value': 'sum'})

    res = orders.merge(grouped_ord, on='order_id', how='left')
    res = res[res['order_status'] == 'delivered']
    res['value'] = res.price + res.freight_value

    whole_sum = sum(res['value'])

    res = res.merge(customers, on='customer_id', how='left')
    res = res.groupby(['customer_state'], as_index=False).apply(lambda g : sum(g['value']) / whole_sum).reset_index(level=0, drop=True)
    res.rename(columns={'customer_state': 'state', None: 'perc'}, inplace=True)
    return res
````

````python
# Проверки

res = task_3(orders, customers, items)
assert np.allclose(res.perc.sum(), 1)
assert np.allclose(res[res.state == "RS"].perc.values[0], 0.055868056429816286)
assert np.allclose(res.sort_values("perc", ascending=True).iloc[0, 1], 0.0005862290943146945)
assert np.allclose(res.sort_values("perc", ascending=False).iloc[0, 1], 0.3741756035817322)
assert len(res) == 27
````

## Задание 4 (10 баллов)

Определите средний чек покупки (добавьте разбивку на стоимость самого заказ и стоимость доставки) и среднее число товаров в заказе

А также определите среднее число покупок на пользователя (обратите внимание на идентификаторы)

````python
%%write_and_run task_4.py

# важно! все зависимости, которые используете (если добавляее новые) в этом классе надо явно продублировать в эту ячейку
import numpy as np
import pandas as pd

def task_4(items, orders, customers):


    res = items.groupby(['order_id'], as_index=False).agg({'price' : 'sum', 'freight_value': 'sum', 'product_id': 'count'})

    avg_ord_price = res.price.mean()
    avg_delivery_price = res.freight_value.mean()
    avg_goods_in_order_cnt = res.product_id.mean()

    res = orders.merge(customers, on='customer_id', how='left')
    gr_by_customer = res.groupby(['customer_unique_id'], as_index=False).agg({'order_id': 'count'})

    avg_orders_per_customer = gr_by_customer.order_id.mean()
    return (avg_ord_price, avg_delivery_price, avg_goods_in_order_cnt, avg_orders_per_customer)
````

## Задание 5 (10 баллов)

Посчитайте CSAT (customer satisfaction - средняя оценка ревью) и отобразите средний CSAT по дням в период с апреля 2017 по апрель 2018 года

Все манипуляции с датой необходимо сделать с помощью datetime и dateutil


Табличка, которая у вас должна получиться:

```
date | csat

"YYYY-MM-DD" | value
```

````python

%%write_and_run task_5.py

# важно! все зависимости, которые используете (если добавляее новые) в этом классе надо явно продублировать в эту ячейку
import numpy as np
import pandas as pd
from dateutil.parser import parse
import datetime

def task_5(reviews):
    res = reviews
    res['date'] = res.review_creation_date.apply(lambda x: parse(x, yearfirst=True))
    res = res.groupby(['date'], as_index=False).agg({'review_score' : 'mean'})
    res.date = res.date.apply(lambda x: x.strftime('%Y-%m-%d'))
    res = res[('2017-04-01' <= res['date']) & (res['date'] <= '2018-04-30')]
    res.rename(columns={'review_score': 'csat'}, inplace=True)
    return res
````

````python
# Проверки
res = task_5(reviews)
assert res.date.min() == '2017-04-01'
assert res.date.max() == '2018-04-30'
assert res.csat.sum() == 1551.8881071384853
assert res[res.date == '2017-07-11'].csat.values[0] == 4.291390728476821
assert res[res.date == '2018-02-09'].csat.values[0] == 3.992156862745098
assert res[res.csat == 3.6814814814814816].date.values[0] == '2018-02-25'
````

## Задание 6 (10 баллов)

Посмотрите, как быстро отвечают пользователи (сделайте аггреграцию по числу дней ответа) и какая средняя оценка

Все манипуляции со временем нужно делать через datetime и dateutil

Табличка, которая у вас должна получиться:

```
days | csat | orders

value | value | value
```

Результаты должны быть отсориртированы по дня по возрастанию

````python
%%write_and_run task_6.py

# важно! все зависимости, которые используете (если добавляее новые) в этом классе надо явно продублировать в эту ячейку
import numpy as np
import pandas as pd
from dateutil.parser import parse
import datetime

def task_6(reviews):
    res = reviews
    res['days'] = (res.review_answer_timestamp.apply(lambda x: parse(x)) - res.review_creation_date.apply(lambda x: parse(x))).apply(lambda x: x.days)
    res.rename(columns={'review_score': 'csat', 'order_id': 'orders'}, inplace=True)
    res = res.groupby(['days'], as_index=False).agg({'csat' : 'mean', 'orders': 'count'})
    return res
````

````python
# Проверки
res = task_6(reviews)
assert res.orders.sum() == 99224
assert np.all(res.days.values == np.sort(res.days.values))
assert len(res) == 214
assert res.days.min() == 0
assert res.days.max() == 518
assert res[res.days == 233].csat.values[0] == 3.0
assert res[res.days == 87].orders.values[0] == 4
````

## Задание 8 (30 баллов)

Определите топ-10 продавцов, у которых больше 100 заказов, отсылающие чаще всего свою посылку в другие регионы (считаются только доставленные заказы)

Чаще всего отсылают = самый большой процент отправленных заказов в другой штат

Табличка, которая у вас должна получиться:

```
seller_id | share

value | value
```

````python
%%write_and_run task_8.py

# важно! все зависимости, которые используете (если добавляее новые) в этом классе надо явно продублировать в эту ячейку
import numpy as np
import pandas as pd

def task_8(orders, items, sellers, customers):
    res = items.merge(orders, on='order_id', how='left')
    res = res.merge(customers, on='customer_id', how='left')
    res = res.merge(sellers, on='seller_id', how='left')
    res = res[res['order_status'] == 'delivered']

    res['another_state'] = (res.customer_state != res.seller_state)
    res = res[['seller_id', 'order_id', 'another_state']]
    res = res.drop_duplicates()
    res = res.groupby(['seller_id'], as_index=False).agg({'another_state' : 'sum', 'order_id' : 'count'})
    res = res[res.order_id > 100]
    res['share'] = res.another_state / res.order_id

    res.sort_values(by ='share', ascending=False, inplace=True)
    res.drop(['order_id', 'another_state'], axis=1, inplace=True)
    res = res[:10]

    return res
````

````python
# Проверки

res = task_8(orders, items, sellers, customers)
assert np.all(res.share.values == np.sort(res.share.values)[::-1])
assert res.share.values[0] == 0.9743589743589743
assert res.share.values[-1] == 0.9356435643564357
assert res.seller_id.values[5] == '1b4c3a6f53068f0b6944d2d005c9fc89'
assert res.seller_id.values[2] == '06a2c3af7b3aee5d69171b0e14f0ee87'
assert res.share.sum() == 9.519118744616716
````
